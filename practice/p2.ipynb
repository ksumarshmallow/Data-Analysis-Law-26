Вам предоставлен датасет, который содержит заявки на патенты в области цифровых технологий.

Датасет содержит следующие признаки:
- `application_id`: Уникальный идентификационный номер патентной заявки.
- `filing_year`: Год подачи заявки в патентное ведомство
- `tech_domain`: Технологическая область изобретения (**ai_ml** = ИИ и машинное обучение, **blockchain** = блокчейн и распределенные реестры, iot = интернет вещей, **quantum** = квантовые вычисления, **cybersecurity** = кибербезопасность).
- `applicant_type`: Тип заявителя (1 = университет/НИИ, 2 = крупная корпорация, 3 = стартап/МСП, 4 = индивидуальный изобретатель).
- `country_origin`: Страна происхождения заявителя (код ISO Alpha-3: **USA**, **CHN**, **DEU**, **KOR**, **JPN**, **GBR**, **RUS**, **IND**).
- `industry_sector`: Отрасль промышленности применения (**software** = программное обеспечение, **hardware** = аппаратное обеспечение, **telecom** = телекоммуникации, **fintech** = финансовые технологии, **healthtech** = медицинские технологии).
- `international_filing`: Подана ли заявка по процедуре РСТ (Международная патентная заявка).
- `prior_examination`: Проводился ли предварительный патентный поиск до подачи заявки.
- `commercial_potential`: Наличие предварительной оценки коммерческого потенциала технологии.
- `market_size_estimate`: Оценка потенциального размера рынка для технологии, в млн долларов США (от 0.1 до 500.0).
- `processing_fee_usd`: Размер патентной пошлины за рассмотрение заявки, в долларах США (от 500.0 до 10,000.0).
- `number_of_claims`: Количество пунктов формулы изобретения (от 1 до 50).
- `number_of_citations`: Количество ссылок на предшествующий уровень техники (от 0 до 200).
- `complexity_level`: Субъективная оценка сложности технического решения (1 = низкая, 2 = средняя, 3 = высокая).
- `examiner_load`: Загруженность патентного эксперта на момент распределения заявки (**Low** = низкая, **Medium** = средняя, **High** = высокая).
- `novelty_score`: Оценка новизны изобретения по результатам первоначальной проверки (**Marginal** = минимальная, **Moderate** = умеренная, **Significant** = значительная, **Breakthrough** = прорывная).
- `grant_probability`: Прогнозируемая вероятность выдачи патента на этапе подачи (**Very Low**, **Low**, **Medium**, **High**, **Very High**).
- `office_action_count`: Количество официальных запросов (требований) от патентного эксперта до принятия решения (от 0 до 10).
